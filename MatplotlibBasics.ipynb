```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
tmpArr = np.array([1,2,3,1,9,4])
```

```python
#line plot
plt.plot(tmpArr)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
#histogram
tmpArr2 = np.array([100,100,200,300,300,300,400,500,500,500,500])
plt.hist(tmpArr2)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
#scatter plot
tmpX = np.array([10, 22, 33, 41, 52])
tmpY = np.array([21, 42, 61, 83, 94])
plt.scatter(tmpX,tmpY)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```
